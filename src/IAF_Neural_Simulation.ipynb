### Importing necessary libraries

```python
import matplotlib.pyplot as plt
```

### Simulation Parameters

- **Spiking Threshold**: The membrane potential threshold at which a neuron spikes.
  
- **Time Constant**: The time constant of the neuron, which determines how quickly the membrane potential changes.
  
- **Conductance**: The conductance of the membrane, affecting the ion flow through it.
  
- **Time Step**: The time increment used in the simulation.
  
- **External Current**: The external current applied to the neuron.
  
- **Simulation Length**: The total duration of the simulation, measured in arbitrary time units.

```python
spiking_threshold = 10
time_constant = 10
conductance = 1
dt = 1
external_current = 15
simulation_length = 100

membrane_potential = [0]
```

### Simulation Loop

```python
for t in range(0, simulation_length, dt):
    v = membrane_potential[-1] + (dt / time_constant) * (-membrane_potential[-1] + conductance * external_current)
    if v > spiking_threshold:
        v = 0
    membrane_potential.append(v)
```

### Plotting the Membrane Potential

```python
plt.plot(membrane_potential)
plt.xlabel("Time [ms]")
plt.ylabel("Voltage")
plt.title(f"Simulation results with External current I = {external_current}")
plt.show()
```
